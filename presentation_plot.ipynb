```python
from mcda.core.matrices import *
from mcda.core.relations import *
from mcda.core.scales import *
from mcda.outranking.srmp import *
```

```python
alternatives = ["B&B", "Moxy", "Ibis"]

criteria = ["Prix (€)", "Qualité (*)", "Lieu"]

scales = {
    "Prix (€)": QuantitativeScale(0, 400, preference_direction=MIN),
    "Qualité (*)": DiscreteQuantitativeScale([1, 2, 3, 4, 5], preference_direction=MAX),
    "Lieu": QualitativeScale(Series([0, 1], index=["périphérie", "centre"]), preference_direction=MAX),
}

dataset = PerformanceTable(
    [
        [282, 3, "centre"],
        [315, 4, "centre"],
        [234, 2, "périphérie"],
    ],
    scales,
    alternatives,
    criteria,
)
```

```python
SRMP.plot_input_data(dataset, annotations=True, annotations_alpha=1, figsize=(4, 3))
```

```python
profiles = PerformanceTable(
    [
        [250, 3, "centre"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations_alpha=1, figsize=(4, 3))
```

```python
profiles = PerformanceTable(
    [
        [250, 3, "centre"],
        [300, 2, "périphérie"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[0, 1], annotations_alpha=1, figsize=(4, 3))
```

```python
criteria_weights = {
    "Prix (€)": 0.6,
    "Qualité (*)": 0.2,
    "Lieu": 0.2,
}
```

```python
SRMP.plot_input_data(dataset, criteria_weights=criteria_weights, profiles=profiles, lexicographic_order=[0, 1], annotations_alpha=1, figsize=(4, 3))
```

```python
profiles_neighbor = PerformanceTable(
    [
        [250, 3, "centre"],
        [300, 3, "périphérie"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles_neighbor, lexicographic_order=[0, 1], annotations_alpha=1, figsize=(4, 3))
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(4, 3))
```

```python
alternatives = ["B&B", "Moxy", "Ibis", "Novotel", "Odalys"]

criteria = ["Price (€)", "Quality (*)", "Place"]

scales = {
    "Price (€)": QuantitativeScale(0, 400, preference_direction=MIN),
    "Quality (*)": DiscreteQuantitativeScale([1, 2, 3, 4, 5], preference_direction=MAX),
    "Place": QualitativeScale(Series([0, 1], index=["suburbs", "center"]), preference_direction=MAX),
}

dataset = PerformanceTable(
    [
        [282, 3, "center"],
        [315, 4, "center"],
        [234, 2, "suburbs"],
        [357, 5, "suburbs"],
        [253, 1, "center"],
    ],
    scales,
    alternatives,
    criteria,
)

profiles = PerformanceTable(
    [
        [250, 4, "center"],
        [300, 3, "suburbs"],
        [350, 2, "suburbs"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations_alpha=1, figsize=(4, 3))
```
