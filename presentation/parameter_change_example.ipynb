```python
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
```

```python
from src.performance_table.normal_performance_table import NormalPerformanceTable
from mcda.outranking.srmp import SRMP
```

```python
A = NormalPerformanceTable(
    [[5, 5, 1], [3, 1, 3], [1, 3, 5]], alternatives=["A", "B", "C"]
)
A.plot()
```

```python
profile = NormalPerformanceTable([[4, 4, 4]], alternatives=["profile"])
weights = dict(zip(range(3), [1, 2, 3]))
Group = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, Group)
print("A = C > B")
```

```python
profile = NormalPerformanceTable([[2, 2, 2]], alternatives=["profile"])
DM = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, DM)
print("C > B > A")
```

```python
profile = NormalPerformanceTable([[4, 2, 2]], alternatives=["profile"])
DM = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, DM)
print("C > B = A")
```

```python
profile = NormalPerformanceTable([[4, 4, 2]], alternatives=["profile"])
DM = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, DM)
print("C = B = A")
```

```python
profile = NormalPerformanceTable([[4, 2, 4]], alternatives=["profile"])
DM = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, DM)
print("C > A > B")
```

```python
profile = NormalPerformanceTable([[4, 4, 4]], alternatives=["profile"])
DM = SRMP(A, weights, profile, [0])
SRMP.plot_input_data(A, DM)
print("C = A > B")
```

Liste de changements de prfrences :
1. B > A -> B = A
2. B = A -> A > B
