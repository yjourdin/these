```python
from mcda.matrices import *
from mcda.relations import *
from mcda.scales import *
from mcda.outranking.srmp import *
from pandas import Series
import matplotlib.pyplot as plt
```

```python
alternatives = ["A", "B", "C"]

criteria = ["Price", "Performance", "Comfort"]

scales = {
    "Price": QuantitativeScale(10000, 50000, preference_direction=MIN),
    "Performance": DiscreteQuantitativeScale(
        [100, 200, 300, 400, 500], preference_direction=MAX
    ),
    "Comfort": QualitativeScale(
        Series([1, 2, 3, 4, 5], index=["*", "a", "**", "b", "***"]), preference_direction=MAX
    ),
}

dataset = PerformanceTable(
    [
        [15000, 100, "*"],
        [25000, 300, "***"],
        [40000, 500, "***"],
    ],
    scales,
    alternatives,
    criteria,
)
```

```python
SRMP.plot_input_data(dataset, annotations=True, annotations_alpha=1, figsize=(8, 6))
plt.savefig("alternatives.pdf")
```

```python
profiles = PerformanceTable(
    [
        [30000, 200, "a"],
        [20000, 400, "b"],
    ],
    scales,
    criteria=criteria,
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations=True, annotations_alpha=0.9, figsize=(8, 6))
plt.savefig("profiles.pdf")
```

```python
SRMP.plot_input_data(
    dataset,
    profiles=profiles,
    lexicographic_order=[1, 0],
    annotations_alpha=1,
    figsize=(8, 6),
)
plt.savefig("lex_order.pdf")
```

```python
criteria_weights = {
    "Price": 0.4,
    "Performance": 0.2,
    "Comfort": 0.4,
}
```

```python
SRMP.plot_input_data(
    dataset,
    criteria_weights=criteria_weights,
    profiles=profiles,
    lexicographic_order=[1, 0],
    annotations_alpha=1,
    figsize=(4, 3),
)
plt.savefig("lex_order.pdf")
```

```python
SRMP.plot_input_data(
    dataset,
    criteria_weights=criteria_weights,
    profiles=profiles,
    # lexicographic_order=[1, 0],
    annotations_alpha=1,
    figsize=(4, 3),
)
plt.savefig("weights.pdf")
```

```python
alternatives = ["B&B", "Moxy", "Ibis", "Novotel", "Odalys"]

criteria = ["Price ()", "Quality (*)", "Place"]

scales = {
    "Price ()": QuantitativeScale(0, 400, preference_direction=MIN),
    "Quality (*)": DiscreteQuantitativeScale([1, 2, 3, 4, 5], preference_direction=MAX),
    "Place": QualitativeScale(
        Series([0, 1], index=["suburbs", "center"]), preference_direction=MAX
    ),
}

dataset = PerformanceTable(
    [
        [282, 3, "center"],
        [315, 4, "center"],
        [234, 2, "suburbs"],
        [357, 5, "suburbs"],
        [253, 1, "center"],
    ],
    scales,
    alternatives,
    criteria,
)

profiles = PerformanceTable(
    [
        [250, 4, "center"],
        [300, 3, "suburbs"],
        [350, 2, "suburbs"],
    ],
    scales,
    criteria=criteria,
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations_alpha=1, figsize=(4, 3))
```
