```python
from mcda.matrices import *
from mcda.relations import *
from mcda.scales import *
from mcda.outranking.srmp import *
from pandas import Series
import matplotlib.pyplot as plt
```

```python
alternatives = ["S1", "S2", "S3"]

criteria = ["Price (€)", "Quality", "Time (days)"]

scales = {
    "Price (€)": QuantitativeScale(2700, 3300, preference_direction=MIN),
    "Quality": QualitativeScale(Series([0, 1, 2, 3], index=["A", "B", "C", "D"]), preference_direction=MIN),
    "Time (days)": DiscreteQuantitativeScale([0, 1, 2, 3, 4, 5, 6], preference_direction=MIN),
}

dataset = PerformanceTable(
    [
        [3075, "A", 5],
        [3013, "D", 1],
        [3175, "A", 3],
    ],
    scales,
    alternatives,
    criteria,
)
```

```python
SRMP.plot_input_data(dataset, annotations=True, annotations_alpha=1, figsize=(8, 6))
plt.savefig("alternatives.pdf")
```

```python
profiles = PerformanceTable(
    [
        [3150, "C", 4],
        [3100, "B", 2],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations=True, annotations_alpha=0.9, figsize=(8, 6))
plt.savefig("profiles.pdf")
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(8, 6))
plt.savefig("lex_order.pdf")
```

```python
criteria_weights = {
    "Price (€)": 0.2,
    "Quality": 0.7,
    "Time (days)": 0.1,
}
```

```python
SRMP.plot_input_data(dataset, criteria_weights=criteria_weights, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(8, 6))
plt.savefig("weights.pdf")
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(4, 3))
```

```python
alternatives = ["B&B", "Moxy", "Ibis", "Novotel", "Odalys"]

criteria = ["Price (€)", "Quality (*)", "Place"]

scales = {
    "Price (€)": QuantitativeScale(0, 400, preference_direction=MIN),
    "Quality (*)": DiscreteQuantitativeScale([1, 2, 3, 4, 5], preference_direction=MAX),
    "Place": QualitativeScale(Series([0, 1], index=["suburbs", "center"]), preference_direction=MAX),
}

dataset = PerformanceTable(
    [
        [282, 3, "center"],
        [315, 4, "center"],
        [234, 2, "suburbs"],
        [357, 5, "suburbs"],
        [253, 1, "center"],
    ],
    scales,
    alternatives,
    criteria,
)

profiles = PerformanceTable(
    [
        [250, 4, "center"],
        [300, 3, "suburbs"],
        [350, 2, "suburbs"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations_alpha=1, figsize=(4, 3))
```
