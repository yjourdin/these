```python
from mcda.matrices import *
from mcda.relations import *
from mcda.scales import *
from mcda.outranking.srmp import *
from pandas import Series
import matplotlib.pyplot as plt
```

```python
alternatives = ["Mark", "Plaza", "Luxus"]

criteria = ["Price (€)", "Quality (*)", "Location"]

scales = {
    "Price (€)": QuantitativeScale(0, 400, preference_direction=MIN),
    "Quality (*)": DiscreteQuantitativeScale([1, 2, 3, 4, 5], preference_direction=MAX),
    "Location": QualitativeScale(Series([0, 1, 2, 3], index=["suburbs", "1", "2", "center"]), preference_direction=MAX),
}

dataset = PerformanceTable(
    [
        [315, 1, "center"],
        [272, 5, "suburbs"],
        [234, 3, "suburbs"],
    ],
    scales,
    alternatives,
    criteria,
)
```

```python
SRMP.plot_input_data(dataset, annotations=True, annotations_alpha=1, figsize=(4, 3))
plt.savefig("alternatives.pdf")
```

```python
profiles = PerformanceTable(
    [
        [280, 4, "2"],
        [300, 2, "1"],
    ],
    scales,
    criteria=criteria
)
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, annotations_alpha=1, figsize=(4, 3))
plt.savefig("profiles.pdf")
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(4, 3))
plt.savefig("lex_order.pdf")
```

```python
criteria_weights = {
    "Price (€)": 0.2,
    "Quality (*)": 0.2,
    "Location": 0.6,
}
```

```python
SRMP.plot_input_data(dataset, criteria_weights=criteria_weights, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(4, 3))
plt.savefig("weights.pdf")
```

```python
SRMP.plot_input_data(dataset, profiles=profiles, lexicographic_order=[1, 0], annotations_alpha=1, figsize=(4, 3))
```
